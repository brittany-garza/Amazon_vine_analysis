```python
import pandas as pd
```

```python
# Creating DF from vine csv
vine_data_to_load = "vine_table.csv"
vine_df = pd.read_csv(vine_data_to_load)
vine_df
```

```python
vine_df.dtypes
```

```python
# 1. Filter total votes to only include votes greater than 20 then make a new dataframe
filtered_total_votes_df = vine_df[(vine_df["total_votes"]  >= 20)]
filtered_total_votes_df
```

```python
# 2. using the dataframe in the previous cell divide the rows where helpful votes by total votes is equal to or great than 50% then make a new dataframe



helpful_total = filtered_total_votes_df.assign(Helpful_Total_votes = (filtered_total_votes_df['helpful_votes']/filtered_total_votes_df['total_votes']))
helpful_total_df = helpful_total[(helpful_total["Helpful_Total_votes"] >= 0.5)]
helpful_total_df
```

```python
# 3. using the dataframe in the previous cell retrieve all rows where a review was part of the vine program then make a new dataframe
# vine == Y
vine_paid_df = helpful_total_df[(helpful_total_df["vine"] == "Y")]
vine_paid_df
```

```python
# 4. using the dataframe in step 2 retrieve all rows where a review was part of the vine program then make a new dataframe
# vine == N
vine_unpaid_df = helpful_total_df[(helpful_total_df["vine"] == "N")]
vine_unpaid_df
```

```python
# 5. Determine the total # of reviews
# what is the number of 5 star reviews?
# Whats is the percentage of 5 star reviews for Y and N?

review_count = len(vine_df['review_id'].unique())
five_star_reviews = vine_df[(vine_df['star_rating'] == 5.0)].count()['star_rating']

paid_five_star_review = vine_paid_df[(vine_paid_df['star_rating'] == 5.0)].count()['review_id']
unpaid_five_star_review = vine_unpaid_df[(vine_unpaid_df['star_rating'] == 5.0)].count()['review_id']

percentage_paid_five_star_reviews = (paid_five_star_review / review_count) * 100
percentage_unpaid_five_star_reviews = (unpaid_five_star_review / review_count) * 100
```

```python
print(f"{review_count} is the total number of reviews")
print(f"{five_star_reviews} is the total number of 5 star reviews")

print(f"The precentage for paid five star reviews is {percentage_paid_five_star_reviews: .2f}%")
print(f"The percentage for unpaid five star reviews is {percentage_unpaid_five_star_reviews: .2f}%")
```
